```python
#HEIGHT OF A BINARY TREE
class Node:
    def __init__(self, info): 
        self.info = info  
        self.left = None  
        self.right = None 
        self.level = None 

    def __str__(self):
        return str(self.info) 

class BinarySearchTree:
    def __init__(self): 
        self.root = None

    def create(self, val):  
        if self.root == None:
            self.root = Node(val)
        else:
            current = self.root
         
            while True:
                if val < current.info:
                    if current.left:
                        current = current.left
                    else:
                        current.left = Node(val)
                        break
                elif val > current.info:
                    if current.right:
                        current = current.right
                    else:
                        current.right = Node(val)
                        break
                else:
                    break

def height(root):
    add = 0
    if root.left:
        add = 1 + height(root.left)
    if root.right:
        add = 1 + height(root.right)
    return add

tree = BinarySearchTree()
arr = [1,2,3,4,5]
t = 5

for i in range(t):
    tree.create(arr[i])

print(height(tree.root))
```

```python
#EQUAL STACKS
def equalStacks(h1, h2, h3):
    #FIND GREATEST COMMON SUM
    sum_1 = sum(h1)
    sum_2 = sum(h2)
    sum_3 = sum(h3)
    while True:
        minimum_height = min(sum_1,sum_2,sum_3)
        #base case
        if(minimum_height == 0):
            return 0
        #general
        if(minimum_height < sum_1):
            popped = h1.pop(0)
            sum_1 -= popped
        if(minimum_height < sum_2):
            popped = h2.pop(0)
            sum_2 -= popped
        if(minimum_height < sum_3):
            popped = h3.pop(0)
            sum_3 -= popped
            
        if(sum_1 == sum_2 == sum_3):
            return minimum_height
                
if __name__ == '__main__':
    
    h1 = [3, 2, 1, 1, 1]
    h2 = [4, 3, 2]
    h3 = [1, 1, 4, 1]

    result = equalStacks(h1, h2, h3)
    print(result)
```
